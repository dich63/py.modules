```python
# -*- coding: utf-8 -*-

from numpy import *
from utils import *

import numpy as np


from tensor.spline4D import *

import tracemalloc

tracemalloc.start()

def tracedump():
    
    current, peak = tracemalloc.get_traced_memory()
    
    print(f"Current memory usage is {round(current / 2**30,3)} GB; Peak was {round(peak / 2**30,3)} GB")
    
    tracemalloc.stop()
    tracemalloc.start()


Na=100
Np=14

X = Y = Z = T = linspace(-2, 2, Np)

nc=int(Np/2)

XX,YY,ZZ,TT = meshgrid(X,Y,Z,T)
FF=random.normal(size=[Np,Np,Np,Np]);

tracedump()

tic('tensor_spline_4D init')
ts=tensor_spline_4D(FF,X,Y,Z,T,full=0)
toc(':')
tracedump()
d=2-1e-7
X2 = Y2 = linspace(-d,d,Na)
XX0,YY0 = meshgrid(X2,Y2);

tic('tensor_spline_4D  [pass 1] make  ')
ff0=ts(XX0,YY0,0.01,0.01);
toc(':')
tracedump()
print('cache_count=',ts.cache_count)

X2 = Y2 = linspace(-0.5, 0.5, Na)


XX2,YY2 = meshgrid(X2,Y2);


tic('tensor_spline_4D [pass 2] make')
ff=ts(XX2,YY2,0.,0.);
toc(':')
print('cache_count=',ts.cache_count)
tracedump()

tracemalloc.stop()
# ---- Graphics.....
# ---- Graphics.....
# ---- Graphics.....
#%matplotlib inline
#%matplotlib widget
import matplotlib.widgets 
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
#plt.ion()
aa=[]
plt.rcParams["figure.figsize"]=7,7
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX[:,:,nc,nc], YY[:,:,nc,nc], FF[:,:,nc,nc], rstride=1, cstride=1,cmap='viridis', edgecolor='none');
aa+=[ax]
plt.show()
plt.pause(0.1);
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX0, YY0, ff0, rstride=1, cstride=1,cmap='viridis', edgecolor='none');
aa+=[ax]
plt.show()
plt.pause(0.1);
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX2, YY2, ff, rstride=1, cstride=1,cmap='viridis', edgecolor='none');
aa+=[ax]
plt.show()
```

***
$\mathbf{\text{Gradient Tree Boosting Algorithm}}$<br>
***
1.&emsp;Initialize model with a constant value $$f_{0}(x) = \textrm{arg min}_{\gamma} \sum \limits _{i=1} ^{N} L(y_{i}, \gamma)$$
2.&emsp;For m = 1 to M:<br>
&emsp;&emsp;(a)&emsp;For $i = 1,2,...,N$ compute<br>
    $$r_{im} = - \displaystyle \Bigg[\frac{\partial L(y_{i}, f(x_{i}))}{\partial f(x_{i})}\Bigg]_{f=f_{m−1}}$$
&emsp;&emsp;(b)&emsp;Fit a regression tree to the targets $r_{im}$ giving terminal regions<br>
&emsp;&emsp;&emsp;&emsp;$R_{jm}, j = 1, 2, . . . , J_{m}.$<br><br>
&emsp;&emsp;(c)&emsp;For $j = 1, 2, . . . , J_{m}$ compute<br>
$$\gamma_{jm} = \underset{\gamma}{\textrm{arg min}} \sum \limits _{x_{i} \in R_{jm}} L(y_{i}, f_{m−1}(x_{i}) + \gamma)$$
<br>
&emsp;&emsp;(d)&emsp;Update $f_{m}(x) = f_{m−1}(x) + \sum _{j=1} ^{J_{m}} \gamma_{jm} I(x \in R_{jm})$<br><br>
3. Output $\hat{f}(x) = f_{M}(x)$
